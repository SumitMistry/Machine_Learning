```python
from pyspark.ml.classification import LogisticRegression
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("LogisticRegressionExample").getOrCreate()
```

# Load training data

```python
# Load training data
training = spark.read.format("libsvm").load("sample_libsvm_data.txt")
```

```python
training.head(5)
```

**Instantiate Logistic Regression object**


```python
lr = LogisticRegression(maxIter=10, regParam=0.3, elasticNetParam=0.8)
```

# Fit the model

```python
lrModel = lr.fit(training)
```

# Print the coefficients and intercept for logistic regression

```python
print("Coefficients: " + str(lrModel.coefficients))
print("Intercept: " + str(lrModel.intercept))
```

```python
training.show()
```

# Explaining label and Features
* label : 0 or 1
* feature : has a vector of all the features that belong to that row

**Instantiate an object of the Logistic Regression**

```python
# Create initial LogisticRegression model
lr = LogisticRegression(labelCol="label", featuresCol="features", maxIter=10)
```

# Splitting the data in training and testing to evaluate the model

```python
train_data,test_data = training.randomSplit([0.70,0.30])
```

```python
train_data. show()
```

```python
test_data.show()
```

# fitting our model to the training data

```python
model = lr.fit(train_data)
```

```python
result = model.evaluate(test_data)
```
